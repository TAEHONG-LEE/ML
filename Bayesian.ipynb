#Iris Bayesian

```python
Iris_df = pd.read_csv('/content/drive/MyDrive/2022/CNU/ML/ML/Iris.csv')
```

```python
Iris_df
```

```python
del Iris_df['Id']

Iris_df
```

```python
from sklearn.preprocessing import LabelEncoder

Iris_df_LE = LabelEncoder().fit_transform(Iris_df[['Species']])
Iris_df_LE
```

```python
Iris_df.Species = Iris_df_LE
Iris_df
```

```python
X = Iris_df.iloc[:, :-1]
y = Iris_df.iloc[:, -1]
```

```python
X.value_counts()
```

```python
y.value_counts()
```

```python
SepalLengthCm = X.SepalLengthCm
SepalWidthCm = X.SepalWidthCm
PetalLengthCm = X.PetalLengthCm
PetalWidthCm = X.PetalWidthCm
```

### 사전확률 조정 x

```python
from sklearn.naive_bayes import GaussianNB as GNB

model = GNB()
model.fit(X, y)

model.predict(X)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(
    SepalLengthCm, SepalWidthCm, 
    c = model.predict(X),
    marker = '.',
    cmap = 'tab10'
)

# c predict에 따라 color를 보여주는 매개변수인데 
# scatter는 한 번에 두 개의 상관관계만 보여준다.
```

```python
result = []

result.append(model.predict(X))
```

### 0번클래스 사전확률 높게 지정 .98 .01 .01

```python
model = GNB(priors = [.98, .01, .01])
model.fit(X, y)
result.append(model.predict(X))
```

### 1번 클래스 사전확률 높게 지정 .01 .98 .01

```python
model = GNB(priors = [.01, .98, .01])
model.fit(X, y)
result.append(model.predict(X))
```

### 2번 클래스 사전확률 높게 지정 .01 .01 .98

```python
model = GNB(priors = [.01, .01, .98])
model.fit(X, y)
result.append(model.predict(X))
```

```python
# 3. 결과 확인
fig, axes = plt.subplots(1, 4, figsize=(20,5))
for i in range(4):
    axes[i].scatter(
        SepalWidthCm, SepalLengthCm, 
        c = result[i], 
        marker = 'o',
        cmap = 'Paired'
        )
```
